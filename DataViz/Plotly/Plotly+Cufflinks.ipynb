### Dados forjados

```python
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'Categoria':['A','B','C'],'Valores':[32,43,50]})
```

```python
df2.head()
```

## Gráficos

## Dispersão

kind = tipo do gráfico

X = Dados a serem exibidos no eixo X

Y = Dados a serem exibidos no eixo Y

mode = Tipo dos marcadores

size = Tamanho dos marcadores

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers',size=10)
```

## Plots de barra

kind = tipo do gráfico

X = Dados a serem exibidos no eixo X

Y = Dados a serem exibidos no eixo Y



```python
df2.iplot(kind='bar',x='Categoria',y='Valores')
```

```python
df.count().iplot(kind='bar')
```

## Boxplots

kind = tipo do gráfico

```python
df.iplot(kind='box')
```

## Superfícies 3D

kind = tipo do gráfico

x, y, z = dimensões para formação do gráfico de superfície

colorscale = escala de cor , no caso red, yellow e blue

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
df3.iplot(kind='surface',colorscale='rdylbu')
```

## Spread

kind = tipo do gráfico

```python
df[['A','B']].iplot(kind='spread')
```

## Histograma

kind = tipo do gráfico

bins = largura das barras do gráfico

```python
df['A'].iplot(kind='hist',bins=15)
```

## Bolha

kind = tipo do gráfico

x = dados a serem exibidos no eixo X

y = dados a serem exibidos no eixo Y

size = tamanha das bolhas, le v

```python
df.iplot(kind='bubble',x='A',y='B',size='C')
```
