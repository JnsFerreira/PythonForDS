## Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

** Importe matplotlib.pyplot como plt e defina% matplotlib inline se você estiver usando o notebook jupyter. Qual comando você usa se você não estiver usando o notebook jupyter? **

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Fora do Jupyter notebook = plt.show
```

## Exercício 1

** Acompanhe estes passos: **
* ** Crie um objeto de figura chamado fig usando plt.figure () **
* ** Use add_axes para adicionar um eixo à tela de figura em [0,0,1,1]. Chame esse novo eixo de "ax". **
* ** Plote (x, y) nesses eixos e defina os rótulos e títulos para corresponder ao gráfico abaixo: **

```python
#Crie um objeto de figura chamado fig usando plt.figure ()
fig = plt.figure()

#Use add_axes para adicionar um eixo à tela de figura em [0,0,1,1]. Chame esse novo eixo de "ax".
ax = fig.add_axes([0,0,1,1])

#Plote (x, y) nesses eixos
ax.plot(x,y)

#Defina os rótulos e títulos para corresponder ao gráfico abaixo
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Title")
```

## Exercício 2
** Crie um objeto de figura e coloque dois eixos sobre ele, ax1 e ax2. Localizado em [0,0,1,1] e [0,2,0,5, .2, .2], respectivamente. **

```python
#Crie um objeto de figura
fig = plt.figure()

#Coloque dois eixos sobre ele, ax1 e ax2. Localizado em [0,0,1,1] e [0,2,0,5, .2, .2], respectivamente
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

** Agora plote (x, y) em ambos os eixos. E chame seu objeto de figura para mostrá-lo. **

```python
#Agora plote (x, y) em ambos os eixos
ax1.plot(x,y)
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_title("Title AX1")

ax2.plot(x,y)
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_title("Title AX2")

#Chame seu objeto de figura para mostrá-lo
fig
```

## Exercício 3

** Crie o gráfico abaixo, adicionando dois eixos a um objeto de figura em [0,0,1,1] e [0,2,0,5, .4, .4] **

```python
#Cria o objeto figura
fig2 = plt.figure()

#Cria dois eixos no objeto figura com as medidas solicitadas
ax3 = fig2.add_axes([0,0,1,1])
ax4 = fig2.add_axes([0.2,0.5,.4,.4])
```

** Agora use x, y e z arrays para recriar o gráfico abaixo. Observe os limites xlimits e y no gráfico inserido: **

```python
#Plota os arrays X e Z
ax3.plot(x,z)
ax3.set_xlabel("X")
ax3.set_ylabel("Z")
ax3.set_title("Title AX3")

#Plota os arrays X e Y, limitando os valores de x e y
ax4.plot(x,y)
ax4.set_xlabel("X")
ax4.set_ylabel("Y")
ax4.set_title("Title AX4")
ax4.set_xlim(20,22) #Limite de X
ax4.set_ylim(30,50) #Limite de Y

#Exibe o resultado
fig2
```

## Exercício 4

** Use plt.subplots (nrows = 1, ncols = 2) para criar o gráfico abaixo. **

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

** Agora trace (x, y) e (x, z) nos eixos. Note a largura de linha e o estilo objetivos. **

```python
axes[0].plot(x,y,color="blue", lw=3, ls='--')
axes[1].plot(x,z,color= "red", lw =3,ls='-' )

fig
```

** Veja se você pode redimensionar o gráfico adicionando o argumento figsize () em plt.subplots () apenas copiando e colando seu código anterior. **

```python
#Redimensiona o gráfico com o argumento figsize()
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))

axes[0].plot(x,y,color="blue", lw=3)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="red", lw=3, ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```
