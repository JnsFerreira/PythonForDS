## Personalizando um Count Plot

set_style =  Define o stilo do gráfico Ex: ticks, whitegrid, darkgrid,etc

despine =  retira por padrão as bordas 'top' e 'right'

figsize = Define o tamanho da figura, sendo os parametro (horizontal, vertical)

```python
sns.set_style('ticks')
plt.figure(figsize=(12,8))
sns.countplot(x='sex', data=tips)
sns.despine()
```

## Personalizando um Linear Model Plot

height = parametro que substituiu o 'size' e tem como função, ajustar a dimensão vertical do gráfico

aspect = parametro que ajusta a dimensão horizontal do gráfico

set_context = personaliza o gráfico de acordo com um contexto passado por parametro

font_scale = Escala da fonte

palette = Define a paleta de cores do gráfico. Consulte mais em : https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
   

```python
sns.lmplot(x='total_bill', y='tip', data=tips, height=3, aspect=6)
```

```python
sns.set_context('poster', font_scale=1.1)
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.set_context('poster', font_scale=0.5)
sns.lmplot(x='total_bill', y='tip', data=tips, hue = 'sex', palette = 'viridis')
```
