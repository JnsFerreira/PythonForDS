```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
#Acessa um dataset ja existente do seaborn
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
tips.describe()
```

```python
sns.distplot(tips['total_bill']) #Opção de tirar o kde sns.distplot(tips['total_bill'], kde=False)
```

```python
sns.jointplot(x= 'total_bill', y = 'tip', data=tips, kind='hex')
```

```python
#Utiliza todas as variáveis numéricas do dataset
sns.pairplot(tips, hue='sex') #hue categoriza de acordo com a coluna passada como parametro
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
#kde + rugplot

sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```
