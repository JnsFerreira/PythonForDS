## Linear Model Plot

x = Coluna do dataframe que será exibida no eixo X

y=  Coluna do dataframe que será exibida no eixo Y

data = dataframe em que os dados buscados

hue = segregação no mesmo gráfico de acordo com a coluna informada

palette = paleta de cores

markers = altera o tipo dos marcadores

scatter_kws = Altera as dimensões do gráfico

col = segregação em outra coluna de acordo com a coluna informada

row = segregação em outra linha de acordo com a coluna informada

aspect / height = Ajusta dimensões do gráfico

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette ='coolwarm', 
           markers =['o','v'], scatter_kws={'s':100}, col='sex', row = 'time',
           aspect =0.6,height =15)
```
