# Plots Categóricos

Importando as bibliotecas e definindo o matplotlib com inline

```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

Carregando os datasets existentes na seaborn 

```python
tips = sns.load_dataset('tips')
```

Algumas informações sobre o dataframe

```python
tips.head()
```

## Barplot

x = Coluna do dataframe que será exibida no eixo X

y = Coluna do dataframe que será exibida no eixo Y

data = Dataframe onde serão buscado os dados

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

## Count Plot

x = Coluna do dataframe que será exibida no eixo X

y = Coluna do dataframe que será exibida no eixo Y

data = Dataframe onde serão buscado os dado

hue = Cria uma nova segmentação de acordo com a coluna informada como parametro

```python
sns.countplot(x='smoker', data=tips, hue='sex')
```

## Box Plot

x = Coluna do dataframe que será exibida no eixo X

y = Coluna do dataframe que será exibida no eixo Y

data = Dataframe onde serão buscado os dado

palette = Paleta de cores

hue = Cria uma nova segmentação de acordo com a coluna informada como parametro



```python
sns.boxplot(x='day', y='total_bill', data = tips, palette = 'rainbow', hue='sex')
```

## Box Plot

data = Dataframe onde serão buscado os dado

palette = Paleta de cores

orient = orientação do gráfico


```python
sns.boxplot(data=tips, palette ='rainbow', orient='h')
```

## Violin Plot

x = Coluna do dataframe que será exibida no eixo X

y = Coluna do dataframe que será exibida no eixo Y

data = Dataframe onde serão buscado os dado

palette = Paleta de cores

hue = Cria uma nova segmentação de acordo com a coluna informada como parametro

split = Unifica os dados

```python
sns.violinplot(x='day', y='tip', data = tips, palette = 'rainbow', hue='sex', split=True)
```

## Strip Plot

x = Coluna do dataframe que será exibida no eixo X

y = Coluna do dataframe que será exibida no eixo Y

data = Dataframe onde serão buscado os dado

palette = Paleta de cores

hue = Cria uma nova segmentação de acordo com a coluna informada como parametro

jitter = Faz com que os dados sejam expandidos horizontalmente conforme o número de ocorrencias no dataframe

dodge = Separa os dados de acordo com as categorias definidas no 'hue'

```python
sns.stripplot(x='day', y='total_bill', data= tips, jitter = True, hue='sex', dodge= True)
```

## Swarm Plot

x = Coluna do dataframe que será exibida no eixo X

y = Coluna do dataframe que será exibida no eixo Y

data = Dataframe onde serão buscado os dado

hue = Cria uma nova segmentação de acordo com a coluna informada como parametro

dodge = Separa os dados de acordo com as categorias definidas no 'hue'

```python
sns.swarmplot(x='day', y='total_bill', data = tips, hue='sex', dodge=True)
```

## Swarm Plot + Violin Plot

```python
sns.swarmplot(x='day', y='total_bill', data = tips, dodge=True, color = 'black')
sns.violinplot(x='day', y='total_bill', data = tips, dodge=True, palette = 'rainbow')
```
