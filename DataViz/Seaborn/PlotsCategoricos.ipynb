```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
sns.countplot(x='smoker', data=tips, hue='sex')
```

```python
sns.boxplot(x='day', y='total_bill', data = tips, palette = 'rainbow', hue='sex')
```

```python
sns.boxplot(data=tips, palette ='rainbow', orient='h')
```

```python
sns.violinplot(x='day', y='tip', data = tips, palette = 'rainbow', hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data= tips, jitter = True, hue='sex', dodge=True)
```

```python
#Pontos so colocados um ao lado dos outros, na medida que aparecem no dataset
sns.swarmplot(x='day', y='total_bill', data = tips, hue='sex', dodge=True)
```

```python
#Combinao entre Swarmplot e Violinplot
sns.swarmplot(x='day', y='total_bill', data = tips, dodge=True, color = 'black')
sns.violinplot(x='day', y='total_bill', data = tips, dodge=True, palette = 'rainbow')
```
