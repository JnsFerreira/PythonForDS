## Pair Grid

pg = objeto que contém PairGrid

sns.PairGrid = Recebe como parametro o dataset a ser utilizado

map = Recebe o tipo de gráfico a ser utilizado

```python
pg = sns.PairGrid(iris)
pg.map(plt.scatter)
```

map_diag = Aplica funções que estão na diagonal

map_upper = Aplica funções que estão acima da diagonal

map_lower = Aplica funções que estão abaixo da diagonal

```python
pg = sns.PairGrid(iris)
pg.map_diag(plt.hist)
pg.map_upper(plt.scatter)
pg.map_lower(sns.kdeplot)
```

Comparando com o pairplot

```python
sns.pairplot(iris, hue='species')
```

## Facet Grid

fg = objeto que recebe o FacetGrid

tips é passado como parametro de dataset a ser usado

col = coluna do dataset que será utilizada como colunas nos gráficos

row = coluna do dataset que será utilizada como linha nos gráficos

hue = segmentação de acordo com a coluna passada como parametro

```python
fg = sns.FacetGrid(tips, col ='time', row='smoker', hue = 'sex')
fg.map(plt.hist, 'total_bill')
```
