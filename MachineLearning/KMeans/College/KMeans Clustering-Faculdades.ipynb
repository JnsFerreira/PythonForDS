### Criação do modelo K-Means

Importando o modelo Kmeans


```python
from sklearn.cluster import KMeans
```

Criando um objeto com a instância do KMeans e definindo o número de clusters = 2

```python
kmeans = KMeans(n_clusters=2)
```

### Treinamento do Modelo

* Observação: Excluindo a coluna Private, para não influenciar no modelo 

```python
kmeans.fit(df.drop('Private',inplace=False, axis=1))
```

## Avaliação

Não há uma maneira perfeita de avaliar o agrupamento se você não tiver os rótulos, no entanto, como temos os rótulos então aproveitamos isso para avaliar nossos clusters.

Criando uma nova coluna para df chamado 'Cluster', onde:

* Escola particular  = 1
* Escola pública = 0

```python
def converter(cluster):
    if cluster == 'Yes':
        return 1 #Particular
    else:
        return 0 #Pública
```

Criando a nova coluna e aplicando a função de acordo com o rótulo 'Private'

```python
df['Cluster'] = df['Private'].apply(converter)
```

```python
df.head()
```

Análisando as métricas com o relatório de classificação.

```python
from sklearn.metrics import classification_report

print(classification_report(df['Cluster'], kmeans.labels_))
```

Resultado consideravel, tendo em vista que o algoritmo está usando apenas os recursos para agrupar as universidades em 2 grupos distintos. 

Poderiamos realizar um breve estudo e prever a quantidade de clusters ideal para uma acurácia maior
